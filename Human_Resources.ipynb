## CREATE TESTING AND TRAINING DATASET 

```python
df.head()
```

putting all categorical data in x_categorical

```python
x_categorical= df.select_dtypes(include='object')
x_categorical
```

we need to change this categorical data to numbers for that we are important library to change categorical data to number

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()

x_categorical_encoded = onehotencoder.fit_transform(x_categorical).toarray()
```

```python
x_categorical_encoded.shape
```

converting this new data to a dataframe

```python
X_cat =pd.DataFrame(x_categorical_encoded)
X_cat 
```

note that we dropped the target 'Atrittion'

```python
X_numerical = df[['Age','DailyRate','DistanceFromHome','Education','EnvironmentSatisfaction','HourlyRate','JobInvolvement','JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked','OverTime','PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']]
X_numerical
```

doing the Concatenation the all categorical and numerical dataframe together

```python
X_all = pd.concat([X_cat,X_numerical],axis=1)
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

X_all is a DataFrame, convert its column names to strings before applying MinMaxScaker

```python
X_all.columns = X_all.columns.astype(str)

X = scaler.fit_transform(X_all)
```

now we are putting our traget column in y variable from our employee dataset

```python
y = df['Attrition']
y
```

# TRAIN AND EVALUATE A LOGISTIC REGRESSION CLASSIFIER

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=101)
```

checking the train test split shape

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
# Testing Set Performance
y_prediction = lr.predict(X_test)
y_prediction
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# we have used .2f for only 2 decimal places
print(f'Accuracy {100 * accuracy_score(y_prediction, y_test):.2f} %')
```

```python
cm = confusion_matrix(y_prediction,y_test)
sns.heatmap(cm,annot=True)
```

```python

print(classification_report(y_test,y_prediction))
```

```python
print(accuracy_score(y_test,y_prediction))
```

# TRAIN AND EVALUATE A RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python

y_predictions = model.predict(X_test)
```

Testing Set Performance

```python
cm = confusion_matrix(y_test,y_predictions)
sns.heatmap(cm,annot=True)
```

```python

print(classification_report(y_predictions,y_test))
```

```python
print(accuracy_score(y_predictions,y_test))
```
